```python
for dataset in gapminder.datasets[:-1]:
    temp = globals()[dataset].reset_index()
    globals()[dataset] = pd.melt(temp, 
                                 id_vars=['Country'], 
                                 value_vars=list(temp.columns[1:]),
                                 value_name=dataset,
                                 var_name='Year')

del temp
```

```python
regions.reset_index(inplace=True)
```

```python
population.head()
```

```python
life_expectancy.head()
```

```python
df_final = (regions.merge(life_expectancy, 
                          on=['Country'], 
                          how='right')
                   .merge(population, 
                          on=['Country', 'Year'], 
                          how='inner')
                   .merge(fertility, 
                          on=['Country', 'Year'], 
                          how='inner'))

df_final.set_index('Year', inplace=True)

df_final = df_final.rename(columns={'life_expectancy': 'Life Expectancy',
                                    'population':      'Population',
                                    'fertility':       'Fertility'})

df_final.head()
```

```python
df_final.to_csv('./data.csv')
```
